## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What I can claim:
- Observed patterns: We can identify distinct, mathematical clusters of pages based on actual, historical metadata (like word count, intent, and engagement rates) currently in our dataset.

- Directional trends: We can show which content archetypes are historically associated with stronger search performance (e.g., top 3 positions) vs. those that sit deep in search results.

- Decision-support: We can build a routing system that suggests tailored editorial or localization workflows for a batch of pages, drastically reducing manual sorting time for content managers.

What I can never claim:
- Causal proof: We cannot prove that changing a page's word count or intent caused a change in rankings, as correlation does not equal causation.

- "Predicting Google": We cannot predict how future search engine algorithm updates will treat these archetypes, nor can we guarantee specific rank increases. Our model only groups pages by their current characteristics and historical behaviors.
